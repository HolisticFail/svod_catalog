```python
#library needed
import pandas as pd
import matplotlib as mlp
import seaborn as sns
```

```python
#netflix csv
df1 = pd.read_csv('netflixData.csv')
df1 = df1.rename(columns={' country':'country'})
df1.columns
```

```python
#delete column not needed: show_id, director, cast, rating, description
df_netflix = df1[['type', 'title', 'country', 'date_added', 'release_year','duration', 'genre']].copy()
df_netflix.head()
```

```python
#df_netflix type repartition
sns.histplot(data=df_netflix, x="type")
```

```python
#df_netflix country repartition
print(df_netflix.country.unique())
print(df_netflix.country.value_counts())
```

```python
#df_netflix duration repartition
print(df_netflix.duration.unique())
display(df_netflix.duration.value_counts())
```
